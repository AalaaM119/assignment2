```python
#Task 1
class Rectangle:
    length=0
    width=0
    def __init__(self,l,w):
        self.length=l
        self.width=w
        
    def perimeter(self):
        perimeter=2*(self.length+self.width)
        return perimeter
    
    def Area(self):
        area=self.length*self.width
        return area
    
    def display(self):
        print('length: ',self.length)
        print('width: ',self.width)
        print('area: ',self.Area())
        print('perimeter',self.perimeter())
```

```python
class Parallelepipede (Rectangle):
    height=0
    def __init__(self,h,l,w):
        super().__init__(self,l,w)
        self.height=h

    def Volume(self):  #The volume of a parallelepiped is equal to the product of its surface area and height.
        v=self.area()*height
        print('Volume : ',v)
        
```

```python
rec=Rectangle(6,3)
print(rec.Area())
rec.display()
```

```python
#Task2
class Person:
    name=' '
    age=0
    
    def __init__(self,name,age):
        self.name=name
        self.age=age
        
    def display(self):
        print('Name: ',self.name)
        print('Age :',self.age)
```

```python
class Student(Person):
    section=''
    def __init__(self,s,n,g):
        super().__init__(n,g)
        self.section=s
        
    def displayStudent(self):
        print('name: ',self.name)
        print('age: ', self.age)
        print('section: ',self.section)
        
```

```python
student=Student('AI','Aalaa',22)
student.displayStudent()
```
